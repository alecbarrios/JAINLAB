```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#import RNA counts for all vitamins at all sequenced timepoints (P0,3,5; or P0,1,2 for folate)
df = pd.read_csv('/Users/alec/Desktop/JAIN LAB/Vitamin Growth Curves/Vitamin-related CRISPR KO screen/Vit_AB_20200821_count_normalized.txt', delimiter='\t')
df.head(10)
```

```python
#parse df to only include control and riboflavin counts 
ribo_df = df.filter(items=['Gene','VitA', 'VitD','VitE','VitG','VitI','VitJ','VitM','VitN','VitP','VitR'])
ribo_df.head(5)
```

```python
#set index to "Gene" item
ribo_df.set_index('Gene', inplace=True)
ribo_df.head(5)
```

```python
ribo_arr = np.array(ribo_df, ndmin=0)
ribo_inf_1 = np.ndarray((77438,5), dtype=float)
ribo_inf_2 = np.ndarray((77438,5), dtype=float)
ribo_arr.shape
```

```python
#Normalize VitD,E,G,I counts by VitA (infection 1 P0) using np
for idx in range(len(ribo_arr[:])):
    for i in range(5):
        ribo_inf_1[idx,i] = np.divide(ribo_arr[idx,i],ribo_arr[idx,0])

#Normalize VitM,N,P,R counts by VitJ (infection 2 P0) using np
for idx in range(len(ribo_arr[:])):
    for i in range(5):
            ribo_inf_2[idx,i] = np.divide(ribo_arr[idx,i+5],ribo_arr[idx,5])
```

```python
 print(ribo_inf_2[:,:])
```

```python
#stack inf1 and inf2 arrays by axis=2 
#(corresponding time/biological replicates between inf1/2 are stacked on z-axis aka axis=2)
ribo_avg_norm = np.stack((ribo_inf_1,ribo_inf_2),axis=2)
ribo_std = np.array((77438,5))

#calculate stdev of corresponding inf1/inf2 samples
ribo_std = np.std(ribo_avg_norm,axis=2)
print(ribo_std[:,:])
```

```python
#Average inf1,2 values (VitA+J/2 = P0
                       #VitD+M/2 = Ribo_P3, 
                       #VitE+N/2 = Ctrl_P3, 
                       #VitG+P/2 = Ribo_P5, 
                       #VitI+R/2 = Ctrl_P5) 
ribo_avg_norm = np.mean(ribo_avg_norm, axis=2)

print(ribo_avg_norm.shape,ribo_avg_norm[:,:])
```

```python
gene_label = np.array(df['Gene'])

passage_label = ['P0', 'Ribo_P3', 'Ctrl_P3', 'Ribo_P5', 'Ctrl_P5']

print(gene_label[:],gene_label.shape,ribo_avg_norm.shape)
```

```python
#label np avg/normalized count data
ribo_labelled = pd.DataFrame(data=ribo_avg_norm, index=gene_label, columns=passage_label)

print(ribo_labelled, ribo_labelled.shape)
```
